```python
import requests
import json
import pandas as pd
import sqlite3
```

```python
# call to REST api
request = requests.get('http://localhost:8080/medish_centrum_randstad/api/netlify?page=1')
```

```python
# check if api works
request.status_code
```

```python
# check type of output
request.headers['content-type']
```

```python
# read data into pandas dataframe
df = pd.read_json(request.text)
```

```python
df.info()
```

```python
# display df
df
```

```python
# column data is of interest. Format is a dictionary of dictionaries

df_part = df['data']
```
